```python
import numpy as np

def f_Gwaves_spectral_RBR(px_in, f_in, dt):
    # Calculate surface gravity wave parameters
    
    n21 = px_in.shape[0]
    n = (n21 - 1) * 2

    # Calculating wave parameters in the gravity (sea-swell) band (between 0.05hz (20 s) and 1Hz (1))
    jj = np.where(np.logical_and(f_in >= 0.05, f_in < 1))

    f = f_in[jj]
    px = px_in[jj]

    m0 = np.sum(px) / (n * dt)
    m1 = np.sum(f * px) / (n * dt)
    m2 = np.sum((f ** 2) * px) / (n * dt)
    m4 = np.sum((f ** 4) * px) / (n * dt)

    hs = 4.0 * np.sqrt(m0)
    hrms = np.sqrt(8 * m0)
    tm02 = np.sqrt(m0 / m2)
    tmean = m0 / m1
    eps = np.sqrt(1.0 - (m2 ** 2) / (m0 * m4))

    # Calculate Hmax and Tpeak using a high cutoff freq.
    ii = np.where(f >= 0.05)
    fc = f[ii]
    pxc = px[ii]
    j = np.argmax(pxc)
    tpeak = 1 / fc[j]

    hmax = 2.0 * hs

    out = {
        'hs': hs,
        'hrms': hrms,
        'hmax': hmax,
        'tm02': tm02,
        'tm01': tmean,
        'tpeak': tpeak,
        'eps': eps
    }

    return out
```

```python
import numpy as np

def f_IGwaves_spectral_RBR(px_in, f_in, dt):
    # Calculate surface gravity wave parameters
    
    n21 = px_in.shape[0]
    n = (n21 - 1) * 2
    out = {}
    
    # Calculating wave parameters in the infragravity band (0.0033 - 0.05Hz)
    jj = np.where(np.logical_and(f_in >= 0.0033, f_in < 0.05))

    f = f_in[jj]
    px = px_in[jj]

    m0 = np.sum(px) / (n * dt)
    m1 = np.sum(f * px) / (n * dt)
    m2 = np.sum((f ** 2) * px) / (n * dt)
    m4 = np.sum((f ** 4) * px) / (n * dt)

    hs = 4.0 * np.sqrt(m0)
    # hs = 3.8 * np.sqrt(m0)
    hrms = np.sqrt(8 * m0)
    tm02 = np.sqrt(m0 / m2)
    tmean = m0 / m1
    eps = np.sqrt(1.0 - (m2 ** 2) / (m0 * m4))

    # Calculate Hmax and Tpeak using a high cutoff freq.
    ii = np.where(f >= 0.05)
    fc = f[ii]
    pxc = px[ii]
    j = np.argmax(pxc)
    tpeak = 1 / fc[j]

    hmax = 2.0 * hs
    out['hs'] = hs
    out['hrms'] = hrms
    out['hmax'] = hmax
    out['tm02'] = tm02
    out['tm01'] = tmean
    out['tpeak'] = tpeak
    out['eps'] = eps

    return out
```
